```python
#importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Churn_Modelling (2).csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#eliminate the row number
#eliminate the surname, customerid
df=df.drop(['RowNumber','CustomerId','Surname'], axis=1, errors='ignore')
```

```python
df.info()
```

```python
#clear the empty value or null values
df=df.dropna(subset=['Age','Gender'])
df.info()
```

```python
df['Geography'].value_counts()
#here dummy encoding can be done as only 3 values are there
```

```python
df['HasCrCard'].value_counts()
#dummy encoding
```

```python
df['NumOfProducts'].value_counts()
#dummy encoding or one hot encoding
```

```python
#dummy encoding on numof products,hascrcard,geography
df=pd.get_dummies(df, columns=['NumOfProducts','HasCrCard','Geography','Gender'])
```

```python
df.head()
```

```python
df.info()
```

```python
#train and test
from sklearn.model_selection import train_test_split
x=df.drop(['Exited'], axis=1)
y=df['Exited']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
#feature scaling in age,balance estimatesalary

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(x_train,y_train)
y_pred=DT.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=200)
RF.fit(x_train,y_train)
y_pred=RF.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier=Sequential()
"""
unis=input+output/2
initializer= how to update the weight
activation function= relu in input and hidden layers
and signoid at the output layer
input_dim= number of features
"""
#input layer
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_dim=17))
#hidden layer 1
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
#hidden layer 2
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
#output layer
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))

```

```python

#COMPILING THE ANN
#binary_crossentropy as it is a binary cross entropy
#if categorical then categorical_crossentropy
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
y_train[0]
#fitting the dataset in the compiled model
classifier.fit(x_train,y_train,batch_size=10,epochs=200)
```

```python
#making prediction and evaluating the model
y_pred=classifier.predict(x_test)
y_pred_ann=(y_pred>0.5)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred_ann)
print(cm)
```

```python
accuracy_score(y_test,y_pred_ann)
```
