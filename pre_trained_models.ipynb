```python
import numpy as np
import pandas as pd
from keras.applications.vgg16 import preprocess_input, decode_predictions
from keras.preprocessing import image
```

```python
from keras.src.applications.vgg16 import VGG16
#load  the vgg16 model and specify the shape of the inout
model=VGG16(weights='imagenet',include_top=True,input_shape=(224,224,3))
```

```python
#loading a image and preprocess it.
img_path='/content/mic.jpg'
img=image.load_img(img_path,target_size=(224,224))
x=image.img_to_array(img)
x=np.expand_dims(x,axis=0)
x=preprocess_input(x)
```

```python
predict=model.predict(x)
```

```python
predict
```

```python
#get the top 5 predictions with class names
top_5=decode_predictions(predict,top=5)[0]
```

```python
for predict in decode_predictions(predict,top=5)[0]:
  print(predict[1],predict[2])
```
